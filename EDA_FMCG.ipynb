```python
df_margin = pd.read_csv('Desktop/Book1.csv',encoding='ISO-8859-1')
df_margin
```

```python
Df_malas = df_margin.loc[df_margin['FirnName']=='Malas Fruit Products']
Df_Manpasand = df_margin.loc[df_margin['FirnName']=='Manpasand']
Df_Guruji = df_margin.loc[df_margin['FirnName']=='Guruji']
Df_HaldiramFoods = df_margin.loc[df_margin['FirnName']=='Haldiram Foods']
Df_Patanjali = df_margin.loc[df_margin['FirnName']=='Patanjali']
Df_Mapro = df_margin.loc[df_margin['FirnName']=='Mapro']
Df_Mapro
```

```python
Df_malas.WholeSale.mean()
Df_malas.Retail.mean()
Malas =Df_malas.Retail.mean() - Df_malas.WholeSale.mean()
Df_Manpasand.WholeSale.mean()
Df_Manpasand.Retail.mean()
Manpasand =Df_Manpasand.Retail.mean() - Df_Manpasand.WholeSale.mean()
Df_Guruji.WholeSale.mean()
Df_Guruji.Retail.mean()
Guruji = Df_Guruji.Retail.mean() - Df_Guruji.WholeSale.mean()
Df_HaldiramFoods.WholeSale.mean()
Df_HaldiramFoods.Retail.mean()
Haldiram =Df_HaldiramFoods.Retail.mean() - Df_HaldiramFoods.WholeSale.mean()
Df_Patanjali.WholeSale.mean()
Df_Patanjali.Retail.mean()
Patanjali =Df_Patanjali.Retail.mean() - Df_Patanjali.WholeSale.mean()
Df_Mapro.Retail.mean()
Mapro =Df_Mapro.Retail.mean() - Df_Mapro.WholeSale.mean()
```

```python
Final_margin= [Malas,Manpasand,Guruji,Haldiram,Patanjali,Mapro]
Final_margin
```

```python
labels = 'Malas', 'Manpasand', 'Guruji', 'Haldiram','Patanjali','Mapro'
sizes = Final_margin
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','lightgreen','lightpink']
explode = (0.1, 0, 0, 0,0,0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```
